```python
import pandas as pd
import openpyxl as op
from openpyxl import load_workbook
```

```python
df = pd.read_excel(r"data\CADASTRO_ISHAOC_2.xlsx")
df = df.drop(['Unidade','CBO', 'N°', 'Conselho','Telefone','Admissão', 'SSHD', 'DDD', 'Escala','TASY', 'MV', 'SENIOR'], axis=1)
df 
```

```python
mapeamento = {
    "Nome Colaborador": "B11",
    "Nascimento": "B12",
    "CPF": "B13",
    "Estado Civil": "B16",
    "Mãe": "B17",
    "Nacionalidade": "B18",
    "Naturalidade": "B19",
    "E-MAIL": "B20",
    "Endereço": "B22",
    "Nº": "B23",
    "Complemento": "B24",
    "Bairro": "B25",
    "Cidade": "B26",
    "UF": "B27",
    "CEP": "B28",
    "Cargo": "B31",
    "Registro ": "B32",
}

valores_fixos = {
    "B30": "CLT",
    "B34": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B35": "SMS",
    "B37": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B38": "SMS"
}
```

```python
for i, row in df.iterrows():
    wb = load_workbook(r"data\SSHD_1604.xlsx")
    ws = wb["SSHD"]
```

```python
for coluna, celula in mapeamento.items():
    ws[celula] = row[coluna]

for coluna, valor in valores_fixos.items():
    ws[celula] = valor
```

```python
wb.save(f"saida_{i}2.xlsx")
```
