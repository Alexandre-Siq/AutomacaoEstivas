```python
mapeamento = {
    "Nome Colaborador": "B11",
    "Nascimento": "B12",
    "CPF": "B13",
    "Estado Civil": "B16",
    "Mãe": "B17",
    "Nacionalidade": "B18",
    "Naturalidade": "B19",
    "E-MAIL": "B20",
    "Endereço": "B22",
    "Nº": "B23",
    "Complemento": "B24",
    "Bairro": "B25",
    "Cidade": "B26",
    "UF": "B27",
    "CEP": "B28",
    "Cargo": "B31",
    "Registro do Funcionário": "B32",
}

valores_fixos = {
    "B30": "CLT",
    "B34": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B35": "SMS",
    "B37": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B38": "SMS"
}
```

```python
# 3. Carrega o arquivo e a aba UMA ÚNICA VEZ antes de começar o loop
caminho_template = r"data\Modelo_Solicitacao_SSHD.xlsx"
wb = load_workbook(caminho_template)
aba_template = wb["SSHD"] # Variável definida com o nome correto!

print("Iniciando a geração das abas...")

# 4. Fluxo de repetição: iterando linha por linha do DataFrame
for index, row in df.iterrows():
    
    # Define o nome da aba (usando o nome do colaborador)
    nome_colaborador = str(row.get('Nome Colaborador', f'Cadastro_{index}'))
    
    # Limpando caracteres inválidos e limitando a 31 caracteres
    nome_aba = "".join(c for c in nome_colaborador if c not in r'/\*?[]:')[:31]
    
    # Duplica a aba "SSHD" matriz mantendo toda a formatação
    nova_aba = wb.copy_worksheet(aba_template)
    nova_aba.title = nome_aba
    
    # 4.1 Preenchendo os dados variáveis do DataFrame
    for coluna_df, celula_excel in mapeamento.items():
        if coluna_df in df.columns and pd.notna(row[coluna_df]):
            nova_aba[celula_excel] = row[coluna_df]
            
    # 4.2 Preenchendo os valores fixos
    for celula_excel, valor in valores_fixos.items():
        nova_aba[celula_excel] = valor

# 5. Opcional: Remover a aba template original que ficou em branco
wb.remove(aba_template)

# 6. Salva o novo arquivo gerado com todas as abas (Salva uma única vez no final)
caminho_saida = r"data\SSHD_PREENCHIDO_ABAS.xlsx"
wb.save(caminho_saida)

print(f"Processamento concluído! Arquivo salvo em: {caminho_saida}")
```

```python
import pandas as pd
import xlwings as xw

print("Iniciando o processamento...")

# 1. Leitura e limpeza (Lembre-se de manter o caminho absoluto)
df = pd.read_excel(r"C:\Users\x0801681\Documents\Automacao\data\1304_e_1604.xlsx")
colunas_remover = ['Unidade', 'CBO', 'Nº', 'Conselho', 'Telefone', 'Admissão', 'SSHD', 'DDD', 'Escala', 'TASY', 'MV', 'SENIOR']
df = df.drop(columns=[c for c in colunas_remover if c in df.columns])

# 2. Dicionários 
mapeamento = {
    "Nome Colaborador": "B11",
    "Data Nascimento": "B12",
    "CPF": "B13",
    "Estado Civil": "B16",
    "Nome Completo da Mãe": "B17",
    "Nacionalidade": "B18",
    "Naturalidade": "B19",
    "E-MAIL": "B20",
    "Endereço": "B22",
    "Nº": "B23",
    "Complemento": "B24",
    "Bairro": "B25",
    "Cidade": "B26",
    "UF": "B27",
    "CEP": "B28",
    "Cargo": "B31", 
    "Registro do Funcionário": "B32"
}

valores_fixos = {
    "B30": "CLT",
    "B34": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B35": "SMS",
    "B37": "COMPLEXO HOSPITALAR DOS ESTIVADORES",
    "B38": "SMS"
}

# 3. Abrindo o Excel em background
app = xw.App(visible=False) 
wb = None 

try:
    caminho_template = r"C:\Users\x0801681\Documents\Automacao\data\Modelo_Solicitacao_SSHD.xlsx"
    wb = app.books.open(caminho_template)
    
    # Esta é a aba "molde"
    aba_template = wb.sheets['SSHD']

    # 4. AQUI ESTÁ O LOOP: Para cada linha na planilha mãe...
    for index, row in df.iterrows():
        
        nome_colaborador = str(row.get('Nome Colaborador', f'Cadastro_{index}'))
        
        # --- NOVIDADE 1: O "dedo-duro" ---
        # Isso vai imprimir no seu terminal quem está sendo processado agora
        print(f"-> Gerando ficha para: {nome_colaborador}") 
        
        nome_aba = "".join(c for c in nome_colaborador if c not in r'/\*?[]:')[:31]
        
        # --- NOVIDADE 2: Ordem correta ---
        # Em vez de colocar depois do molde, colocamos depois da ÚLTIMA aba existente (wb.sheets[-1])
        nova_aba = aba_template.copy(after=wb.sheets[-1], name=nome_aba)
        
        # 4.1 Preenchendo os dados variáveis 
        for coluna_df, celula_excel in mapeamento.items():
            if coluna_df in df.columns and pd.notna(row[coluna_df]):
                nova_aba.range(celula_excel).value = row[coluna_df]
        
        # 4.1.2 Inserindo valores fixos para os campos de lista suspensa
        # Atribui "Não informado" sempre, ignorando o que estiver na planilha mãe
        nova_aba.range('B14').value = "Não informado"  # Campo Gênero
        nova_aba.range('B15').value = "Não Informado" # Campo Orientação Sexual
                
        # 4.2 Preenchendo os valores fixos
        for celula_excel, valor in valores_fixos.items():
            nova_aba.range(celula_excel).value = valor

    # 5. A SOLUÇÃO: Deleta a aba "molde" que ficou em branco no início!
    aba_template.delete()
    
    # --- NOVIDADE 3: Foco na primeira pessoa ---
    # Deixa a aba do primeiro funcionário ativada ao abrir o Excel
    wb.sheets[0].activate() 

    # 6. Salva o arquivo final
    caminho_saida = r"C:\Users\x0801681\Documents\Automacao\data\SSHD_PREENCHIDO_PERFEITO.xlsx"
    wb.save(caminho_saida)
    print(f"Sucesso! Arquivo salvo sem erros em: {caminho_saida}")

except Exception as e:
    print(f"Ocorreu um erro: {e}")

finally:
    # 7. FECHAMENTO SEGURO
    if wb is not None: 
        wb.close()
    app.quit()
```
